```python
import numpy as np
```

```python
X = 2 * np.random.rand(100,1)
y = 4 + 3 * X + np.random.randn(100,1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Step 1: Prepare the Data
data = {
    'mouse_movements': [150, 10, 200, 5],
    'avg_mouse_speed': [2.5, 1.0, 3.0, 0.8],
    'total_clicks': [30, 5, 50, 2],
    'avg_click_interval': [0.5, 0.05, 0.4, 0.03],
    'num_scrolls': [20, 1, 25, 0],
    'avg_scroll_depth': [500, 20, 600, 10],
    'typing_speed': [40, 100, 45, 120],
    'num_keystrokes': [200, 500, 220, 300],
    'form_fill_time': [120, 10, 100, 8],
    'num_corrections': [5, 0, 3, 1],
    'session_duration': [600, 10, 800, 5],
    'avg_action_delay': [0.1, 0.01, 0.2, 0.02],
    'label': ['human', 'bot', 'human', 'bot']
}

df = pd.DataFrame(data)

# Encode labels
df['label'] = df['label'].map({'human': 1, 'bot': 0})

# Features and target variable
X = df.drop('label', axis=1)
y = df['label']

# Step 2: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train the Model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Step 4: Evaluate the Model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(report)
```
